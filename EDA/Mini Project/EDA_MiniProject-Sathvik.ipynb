```python
# Convert 'Value' to Float after getting rid of currency symbol and suffix.
c=[]
for i in df.Value:
    if i[-1]=='M':
            m = float(i[1:-1])
            f = m*1000000
            c.append(f)
    elif i[-1]=='K':
            k = float(i[1:-1])
            j = k*1000
            c.append(j)
    else:
        j=float(i[1:])
        c.append(j)

m=np.array(c)
df.Value=m
df['Value'].fillna(df['Value'].mean(),inplace=True)
```

```python
df.Value.isnull().sum()
```

```python
# Convert to Float after getting rid of currency symbol and suffix.
c=[]
for i in df.Wage:
    if i[-1]=='M':
            m = float(i[1:-1])
            f = m*1000000
            c.append(f)
    elif i[-1]=='K':
            k = float(i[1:-1])
            j = k*1000
            c.append(j)
    else:
        j=float(i[1:])
        c.append(j)

m=np.array(c)
df.Wage=m
df['Wage'].fillna(df['Wage'].mean(),inplace=True)
```

```python
df.Wage
```

```python
# Convert 'Joined' to int with only year
p=[]
for i in df.Joined:
    if isinstance(i,float)!=True:
        v = int(i[-4:])
        p.append(v)
        
    else:
        p.append(i)

m = np.array(p)
df.Joined = m
```

```python
df.Joined.fillna(df.Joined.mean(),inplace=True)
```

```python
# Convert 'Contract Valid Until' to datetime type
y =[]
import datetime
for i in df['Contract Valid Until']:
    if isinstance(i,float)==False:
        if len(i) == 4:
            d = datetime.datetime.strptime(i, '%Y')
            y.append(d)
        else:
            d = datetime.datetime.strptime(i, '%b %d, %Y')
            y.append(d)
    else:
        y.append(i)
            
o = np.array(y)
print(o)
df['Contract Valid Until']=o
df['Contract Valid Until'].fillna(df['Contract Valid Until'].mean(),inplace=True)
```

```python
# Convert 'Height' to float type
j=[]
for i in df.Height:
    if isinstance(i,float)==False:
        o = i.replace("'",".")
        k = float(o)
        l=k/12
        j.append(l)
    else:
        j.append(i)
m=np.array(j)
print(m)
df.Height=m
df.Height.fillna(df.Height.mean(),inplace=True)
```

```python
# Remove the suffix and convert to float in 'Weight'

p=[]
for i in df.Weight:
    if isinstance(i,float)!=True:
        v = float(i[:3])
        p.append(v)        
    else:
        p.append(i)

m = np.array(p)
df.Weight=m
df.Weight.fillna(df.Weight.mean(),inplace=True)
```

```python
c=[]
for i in df['Release Clause']:
    if isinstance(i,float)==False:
        if i[-1]=='M':
            m = float(i[1:-1])
            f = m*1000000
            c.append(f)
        elif i[-1]=='K':
            k = float(i[1:-1])
            j = k*1000
            c.append(j)
        else:
            j=float(i[1:])
            c.append(j)
    else:
        c.append(i)
m=np.array(c)
df['Release Clause']=m
df['Release Clause'].fillna(df['Release Clause'].mean(),inplace=True)
```
