1.	How many matches we've got in the dataset?

2.	How many seasons we've got in the dataset?

```python
# Total number of seasons
print("The number of seasons are", data1.season.value_counts().count())
```

3.	Which Team had won by maximum runs?

```python
# Team that won by maximum runs
for i in range(len(data1)):
    if data1['win_by_runs'][i] == data1['win_by_runs'].max():
        print("The team that won by maximum runs is", data1['winner'][i], "by", data1.win_by_runs.max(), "runs")
        

data1.sort_values(by='win_by_runs').plot(kind='line',x= 'winner',y='win_by_runs')
plt.xticks(rotation=30,horizontalalignment='right')
plt.show()
```

4.	Which Team had won by maximum wicket?

```python
# Team that won by maximum wickets
print("The teams that won by maximum wickets are\n")
for i in range(len(data1)):
    if data1['win_by_wickets'][i] == data1['win_by_wickets'].max():
        print(data1['winner'][i], 'by', data1['win_by_wickets'].max(), 'wickets')
```

5.	Which Team had won by closest Margin (minimum runs)?

```python
# Team that won by the closest margin runs
print("The teams that won by closest Margin are\n")
for i in range(len(data1)):
        if data1['win_by_wickets'][i] == 0:
            if data1['win_by_runs'][i] == 1:
                    print(data1['winner'][i], 'won by', data1['win_by_runs'][i], 'runs')
```

6.	Which Team had won by minimum wicket?

```python
# Team that won by minimum wickets
for i in range(len(data1)):
        if data1['win_by_runs'][i] == 0:
            if data1['win_by_wickets'][i] == 1:
                    print('The', data1['winner'][i], 'won by', data1['win_by_wickets'][i], 'wicket')
```

7.Which Season had most number of matches

```python
# Season with most number of matches
matches = data1.season.value_counts()
print("The season with most number of matches is", matches.idxmax(),"with",data1.season.value_counts().max(),"matches in total")
data1['season'].value_counts().plot(kind = 'bar',title = 'Number of matches played in each season')
plt.xlabel('Season')
plt.ylabel('Number of matches')
plt.show()
```

8.	Which IPL Team is more successful?

```python
# Which IPL team is the most successful
a = data1.winner.value_counts()
print("The most successful team is", a.idxmax(), "with", data1['winner'].value_counts().max(), "wins")
```

9.	Has Toss-winning helped in winning matches?

```python
tw_mw = data1[data1.toss_winner == data1.winner]
tw_nmw = data1[data1.toss_winner != data1.winner]
plt.pie([len(tw_mw),len(tw_nmw)],radius=0.8,autopct='%0.1f',shadow=True)
plt.legend(('Match Winner is the Toss Winner','Match winner is not the Toss winner'))
plt.show()
print("From the above chart, it is clear that the ratio of match-win and toss-win is high.")
print("Hence, the toss-winning has helped in winning matches")
```
